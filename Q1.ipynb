<h2>Part A: Describe the Structure of the Network</h2>

The network is a 9 layer network. It has 7 hidden layers with 1 input layer and 1 output layer. Each of the 7 hidden layers have 20 neurons. The input layer has 2 inputs while the output layer has 3 outputs representing the R,G,B values. 

Note: each of the 7 hidden layers are of defined as 'Fully Connected' layers i.e. a layer of neurons that perform weighted additions of all inputs (activations on layer below). 

The activation function for each of the hidden layers is an ReLU i.e. rectified linear units

<h2> Part B: What does loss mean here? </h2>

Source from GitHub source code: https://github.com/karpathy/convnetjs/blob/master/src/convnet_layers_loss.js 

The loss function depicts the difference between the correct training values array provided by the user and the input value x at each hidden regression layer. In other words, it measures the difference between the predicted R,G,B value for each pixel and the correct value for each of those pixels (from user input). 

The loss function used here is an example of a Mean Squared Error loss function depicted by the formula: "sum_i(||x_i - y_i||^2)" (as explained in the original GitHub source code). Note: this is implemented in the regression layer. 


<h2> Part C: Plot the loss over time </h2>

```python
#We let the program run and measured the loss value every 500 iterations or so.
## We then stored the loss values at each of those stages in an array and plotted them
## Values of loss every 500 iterations are as follows: [0.01365, 0.00966, 0.00825, 0.00757, 0.00705, 0.00666, 
## 0.00634, 0.00612, 0.00574, 0.00587]

import numpy as np
import matplotlib.pyplot as plt


losses = {500: 0.01365, 1000: 0.00966, 1500: 0.00825, 2000: 0.00757, 2500: 0.00705, 3000: 0.00666, 3500: 0.00634, 4000: 0.00612, 4500: 0.00574, 5000: 0.00587}
lossList = sorted(losses.keys())
sorted_vals = [losses[k] for k in lossList]

print lossList
print sorted_vals

fig = plt.figure()
plt.plot(lossList, sorted_vals)

plt.title('Plotting loss at every 500 iterations with learning rate = 0.01')
plt.xlabel('Iteration Count')
plt.ylabel('Loss Value')
plt.show()
```

<h2> Part D: Halving Learning Rate every 500 Iterations </h2>

```python
#We halved the learning rate every 1000 iterations. The loss values at every 1000 iterations were as follows:
## [0.00946, 0.00566, 0.00432, 0.00381, 0.00353]

loss_half = {1000: 0.00946, 2000: 0.00566, 3000: 0.00432, 4000: 0.00381, 5000: 0.00353}
lossList_half = sorted(loss_half.keys())
lossList_half_vals = [loss_half[k] for k in loss_half]

fig = plt.figure()
plt.plot(lossList, sorted_vals, label="With learning rate = 0.01")
plt.plot(lossList_half, lossList_half_vals, label="With learning rate halved every 500 iterations")

plt.legend()

plt.title('Plotting loss at every 500 iterations with learning rate = 0.01')
plt.xlabel('Iteration Count')
plt.ylabel('Loss Value')
plt.show()
```

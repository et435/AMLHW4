```python
import pandas as pd
import sklearn
import PIL
from PIL import Image
import numpy
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
import random
from sklearn.neighbors import KNeighborsClassifier
```

```python
inputMona = Image.open('MonaLisa.jpg')
outputMona = Image.open('MonaLisa.jpg')
inputPixles = inputMona.load()
outputPixles = outputMona.load()
sizeX,sizeY = inputMona.size
print sizeX
print sizeY
print sizeX * sizeY
r,g,b = inputPixles[40,40]
print r,g,b
```

```python
inputR = numpy.zeros(shape=(sizeX,sizeY))
inputG = numpy.zeros(shape=(sizeX,sizeY))
inputB = numpy.zeros(shape=(sizeX,sizeY))

for x in xrange(0,sizeX):
    for y in xrange(0,sizeY):
        r,g,b = inputPixles[x,y]
        inputR[x,y] = r
        inputG[x,y] = g
        inputB[x,y] = b
```

```python
# processedR = (inputR/(255))
# processedG = (inputG/(255))
# processedB = (inputB/(255))

processedR = inputR
processedG = inputG
processedB = inputB
```

```python
coordinateList = []
for coordinate in xrange(0,10000):
    x = random.randint(0,sizeX-1)
    y = random.randint(0,sizeY-1)
    coordinateList.append([x,y])

```

```python
chosenR = []
chosenG = []
chosenB = []

for coordinate in coordinateList:
    x,y = coordinate
    chosenR.append(processedR[x,y])
    chosenG.append(processedG[x,y])
    chosenB.append(processedB[x,y])

print chosenB
```

```python
clfR = RandomForestClassifier()
clfG = RandomForestClassifier()
clfB = RandomForestClassifier()

clfR.fit(coordinateList,chosenR)
clfG.fit(coordinateList,chosenG)
clfB.fit(coordinateList,chosenB)
```

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.RandomForestClassifier.html

```python
for x in xrange(0,sizeX):
    for y in xrange(0,sizeY):
        r = clfR.predict([[x,y]])
        g = clfG.predict([[x,y]])
        b = clfB.predict([[x,y]])
        rgb = (r,g,b)
        outputPixles[x,y] = rgb  # Set the RGBA Value of the image (tuple)
```
